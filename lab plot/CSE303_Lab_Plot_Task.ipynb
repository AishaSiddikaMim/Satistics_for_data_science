# Exercise 1:
Get total profit of all months and show line plot with the following Style properties
Generated line plot must include following Style properties: â€“

Line Style dotted and Line-color should be red

Show legend at the lower right location.

X label name = Month Number

Y label name = Sold units number

Add a circle marker.

Line marker color as read

```python
# Code Here

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('company_sales_data.csv')

axis = df.plot.line('month_number', 'total_profit', color = 'red', linestyle = '--', linewidth = 4, marker = 'o', ms = 7 , mfc = 'k', label = "Profit data of last year",)
axis.set_xticks(df.month_number) 
axis.set_title('Company Sales Data of last year')
axis.set_xlabel('Month Number')
axis.set_ylabel('Profit in dollar')

plt.show(['Company sales data of last year'])
```

# Exercise 2: 
Read all product sales data and show it  using a multiline plot
Display the number of units sold per month for each product using multiline plots. (i.e., Separate Plotline for each product ).

The graph should look like this.

image.png

```python
# Code Here

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('company_sales_data.csv')

axis1 = plt.subplot()

axis1.plot(df.month_number,df.facecream,'o-',label="Face cream Sales Data")
axis1.plot(df.month_number,df.facewash,'o-',label="Facewash Sales Data")
axis1.plot(df.month_number,df.toothpaste,'o-',label="ToothPaste Sales Data")
axis1.plot(df.month_number,df.bathingsoap,'o-',label="Bathingsoap Sales Data")
axis1.plot(df.month_number,df.moisturizer,'o-',label="Moisturizer Sales Data")
axis1.plot(df.month_number,df.shampoo,'o-',label="Shampoo Sales Data")
axis1.set_xticks(df.month_number)

plt.xlabel("Months Number")
plt.ylabel("Sales units in number")
plt.title('sales data')

plt.legend()
plt.show()
```

# Exercise 3: 
Read face cream and facewash product sales data and show it using the bar chart
The bar chart should display the number of units sold per month for each product. Add a separate bar for each product in the same chart.

```python
# Code Here

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('company_sales_data.csv')

axis2 = df.plot.bar(x = 'month_number', y = ['facewash','facecream'], color = ['steelblue', 'red'])

plt.xlabel('Months Number')
plt.ylabel('Sales units in number')
plt.title('Facewash and facecream sales data')
plt.grid(linestyle = '--')

plt.legend()
plt.show()
```

# Exercise 4:
Calculate total sale data for last year for each product and show it using a Pie chart
Note: In Pie chart display Number of units sold per year for each product in percentage.

The Pie chart should look like this.

image.png

```python
# Code Here

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('company_sales_data.csv')

list = df ['month_number'].tolist ()
label = ['FaceCream', 'FaseWash', 'ToothPaste','Bathingsoap','Shampoo', 'Moisturizer']
pieChart =  [df ['facecream']. sum(), df ['facewash'].sum(), df ['toothpaste'].sum(),df ['bathingsoap'].sum(), df ['shampoo'].sum(), df ['moisturizer'].sum()]
plt.axis("equal")

plt.pie(pieChart, autopct='%1.1f%%', labels = label)
plt.title('Sales data')

plt.legend(loc = 'lower right')
plt.show()
```

# Exercise 5:
Plot Sine and Cosine Graph.

The plot graph should look like this.

image.png

```python
# Code Here

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-np.pi, np.pi, 256, endpoint=True)
y = np.cos(x)
y1 = np.sin(x)

plt.plot(x, y, color='blue',label="Cosine")
plt.plot(x, y1, color='green',label="Sin")

plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title('Sin and Cosine Graph') 

plt.legend()
plt.show()
```
